```python
import optuna
from HPO_utils import make_objective
from Parser import parse_dataset, split_dataset
```

```python
drain_params = {
    'st': [0.4, 0.5, 0.6],
    'depth': [4, 5, 6]
}

for st in drain_params['st']:
    for depth in drain_params['depth']:
        dataset_name = f"Linux_drain_st{st}_depth{depth}"
        print(f"Parsing dataset with Drain st={st} depth={depth}...")
        parse_dataset(
            st=st,
            depth=depth
        )
        split_dataset(
            dataset_dir=f"datasets/{dataset_name}",
            random_seed=42
        )

study = optuna.create_study(
    direction="maximize",
    sampler=optuna.samplers.TPESampler(seed=42)
)

study.optimize(make_objective("Linux", True), n_trials=50)

print("Best TOP-5 accuracy:", study.best_value)
print("Best params:", study.best_params)
```
